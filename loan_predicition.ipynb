```python
#Coverting the variable 'Dependents' to numeric, pls Note '3+'
round_  = range(0,613)
for i in round_:
    if loan['Dependents'][i] == '1':
        loan['Dependents'][i] = 1
    elif loan['Dependents'][i] == '2':
        loan['Dependents'][i] = 2
    elif loan['Dependents'][i] == '3+':
        loan['Dependents'][i] = 3
loan['Dependents'] = pd.to_numeric(loan['Dependents'])
```

```python
#since this is a home-loan
#checking for my hypothesis
#1) The Higher the Income, The Higher the Chances to get a loan
#2) The Higher the Coapplicantincome, the higher chance to get a loan
#3) Urban are more likely to get a loan
#4) Graduate have higher chances to get a loan
#5) Self-Employed are less likely to get a loan than their opposite
#6) A good credit_history is likely to get a loan
#7)Married are likely to get a loan
```

```python
# 'The Higher the Income, The Higher the Chances to get a loan'
def hypothesis_check(column):
    for i in round_:
        value = loan[column][i]
        mean = loan[column].mean()
        std = loan[column].std()
        nth = loan[column].count()
        ans = (mean-value)/(std/np.sqrt(nth))
        return ans
```

```python
#creatin a new variable to identify the percentage of the loan to be paid per month
#but first convert 'LoanAmount' to Thousand
for i in round_:
    loan['LoanAmount'][i] = loan['LoanAmount'][i]*1000
```

```python
loan
```

```python
#just trying to create new columns that will be used
loan['Total_Income'] = 1
loan['Payment_per_month'] = 2
loan['Percentage_of_ApplicantIncome'] = 3
```

```python
#sumin up the Applicant and CoapplicantIncome to get a Total Income
for i in round_:
    loan['Total_Income'][i] = loan['ApplicantIncome'][i]+loan['CoapplicantIncome'][i]
```

```python
#NB: Loan_Amount_Term is in Months, so dividing the LoanAmount with it's term, we can get the figure that 
#The Applicant and the coapplicant(if any) will be paying monthly
for i in round_:
    loan['Payment_per_month'][i] = round(loan['LoanAmount'][i]/loan['Loan_Amount_Term'][i], 2)
```

```python
#Having gotten the amount that the applicant will pay monthly, we now divide the amount with he's
#Income to get a view of how many percent of the applicant income will be used to pay up he's Loan during 
#This Loan_Amount_Term
for i in round_:
    loan['Percentage_of_ApplicantIncome'][i] = round(loan['Payment_per_month'][i]/loan['Total_Income'][i]*100, 2)
```

```python
#Using same method
test_target = strat_test_set['Loan_Status']
strat_test_set = strat_test_set.drop(['Loan_ID','Loan_Status'], axis=1)
```

```python
#Data Transformation using MinMax method...4mular is value-min/max-min*1 for each columns
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
piper = Pipeline([
    ('MinMax', MinMaxScaler())
])
loan_prepared = piper.fit_transform(strat_train_set)
```

```python
loan_prepared
```

```python
#Training the Dataset Using Logistic Regression
from sklearn.linear_model import LogisticRegression
LogReg = LogisticRegression()
LogReg.fit(loan_prepared,train_target)
```

```python
test_data = piper.fit_transform(strat_test_set)
test_label = test_target
logreg_prediction = LogReg.predict(test_data)
#using classification report to evulate the Algorithm performance
from sklearn.metrics import classification_report
print(classification_report(test_label, logreg_prediction))
```
